## Creat new column diagonsis 'Classification of Disease'
- from the column "Classification of Disease" written many words including Benign and Malignant 
- from these we made it that take Benign and Malignant among words and than if not found either of the word take As NA
- finaly we filled missing value using mode 


```python
# Function to filter values
def filter_disease_classification(value):
    if pd.isna(value):
        return None
    value = value.lower()
    if 'benign' in value:
        return 'Benign'
    elif 'malignant' in value:
        return 'Malignant'
    else:
        return None

# Apply the filter function to create the 'diagnosis' column
df['Severity_level_classification'] = df['Classification of Disease'].apply(filter_disease_classification)

# Fill NA values in 'diagnosis' with the mode of the column
mode_value = df['Severity_level_classification'].mode()[0]
df['Severity_level_classification'].fillna(mode_value, inplace=True)
```

## Delete unessessory columns

```python
df.drop(['Patient Name'],axis=1, inplace=True)
df.drop(['By'],axis=1, inplace=True)
df.drop(['Classification of Disease Category'],axis=1, inplace=True)
df.drop(['#'],axis=1, inplace=True)
```

## Stages of Severity_level
- Benign: is either stage 1 or stage 2
- Malignant: is stage 3 or stage 4 

```python
# Function to determine the stage based on the classification of disease
def determine_stage(classification):
    if 'Benign neoplasm' in classification:
        return 'stage1'
    elif 'Malignant neoplasm' in classification:
        return 'stage3'
    elif 'Malignant' in classification:
        return 'stage4'
    else:
        return 'stage2'

# Apply the function to the 'Classification of Disease' column to create the new 'Stage' column
df['Stage'] = df['Classification of Disease'].apply(determine_stage)
```
